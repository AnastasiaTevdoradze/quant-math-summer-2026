```python
import numpy as np 
import matplotlib.pyplot as plt

x = np.linspace(-3, 5, 1000)

def f(x): 
    return x**3 - 3*x**2 - 9*x + 5
```

```python
h = 1e-5

y   = f(x)
fp  = np.array([(f(xi + h) - f(xi)) / h for xi in x])   # f'(x)
fpp = np.array([(f(xi + h) - 2*f(xi) + f(xi - h)) / h**2 for xi in x])  # f''(x)
```

```python
fig, axes = plt.subplots(3, 1, figsize=(10, 10))
ax = axes[0]
ax.plot(x, y, color='steelblue', linewidth=2)
ax.axhline(y=0, color='black', linewidth=0.7)
ax.axvline(x=0, color='black', linewidth=0.7)
ax.set_title("f(x) = x³ - 3x² - 9x + 5", fontsize=12)
ax.set_ylabel("f(x)")
ax.grid(True, alpha=0.3)

ax = axes[1]
ax.plot(x, fp, color='darkgreen', linewidth=2)
ax.axhline(y=0, color='black', linewidth=0.7)
ax.set_title("f'(x) — first derivative", fontsize=12)
ax.set_ylabel("f'(x)")
ax.grid(True, alpha=0.3)

# shade green where f' > 0 (function increasing)
ax.fill_between(x, fp, 0, where=(fp > 0), 
                color='green', alpha=0.15, label="f' > 0 (increasing)")

# shade red where f' < 0 (function decreasing)
ax.fill_between(x, fp, 0, where=(fp < 0), 
                color='red', alpha=0.15, label="f' < 0 (decreasing)")

ax.legend(fontsize=9)

ax = axes[2]
ax.plot(x, fpp, color='darkorange', linewidth=2)
ax.axhline(y=0, color='black', linewidth=0.7)
ax.set_title("f''(x) — second derivative", fontsize=12)
ax.set_ylabel("f''(x)")
ax.set_xlabel("x")
ax.grid(True, alpha=0.3)

# shade blue where f'' > 0 (concave up)
ax.fill_between(x, fpp, 0, where=(fpp > 0),
                color='blue', alpha=0.15, label="f'' > 0 (concave up)")

# shade orange where f'' < 0 (concave down)
ax.fill_between(x, fpp, 0, where=(fpp < 0),
                color='orange', alpha=0.15, label="f'' < 0 (concave down)")

ax.legend(fontsize=9)
```

```python
plt.suptitle("Sign chart visualizer - f(x), f'(x), f''(x)", fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```
