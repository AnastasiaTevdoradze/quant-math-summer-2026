```python
import numpy as np 
import matplotlib.pyplot as plt
fig, axes = plt.subplots(2, 3, figsize=(14, 8))
x = np.linspace(-3, 3, 500)
functions = [
    (x**2, "f(x) = x^2", "steelblue"),
    (x**3, "f(x) = x^3", "crimson"),
    (np.sin(x), "f(x) = sin(x)", "seagreen"),
    (np.exp(x), "f(x) = e^x", "darkorange"),
    (1/x, "f(x) = 1/x", "purple"),
]
for i, (y, title, color) in enumerate(functions):
    row = i //3
    col = i % 3
    ax = axes[row ,col]

    ax.plot(x, y, color=color, linewidth=2)
    ax.axhline(y=0, color='black', linewidth=0.7)
    ax.axvline(x=0, color='black', linewidth=0.7)
    ax.set_title(title, fontsize=12)
    ax.set_xlabel("x", fontsize=12)
    ax.grid(True, alpha=0.3)
    ax.set_ylim(-5, 5) #all subplots to have the same y-axis limits

axes[1, 2].set_visible(False)
plt.suptitle("Five Core Functions", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()

    #for each function in my list, figure out which subplot position it belongs to, 
    #grab that subplot, and draw the curve with its title and color on it
 

```

```python
def numerical_deriv(f, x, h=1e-5):
    return (f(x + h) - f(x)) / h


# test 1: f(x) = x^2, analytical derivative at x=3 is 2x = 6
f1 = lambda x: x**2
numerical = numerical_deriv(f1, 3)
analytic = 2 * 3
print(f"x^2 at x = 3 | numerical: {numerical:.6f} | analytic: {analytic:.6f} | error: {abs(numerical - analytic):.2e}")

# test 2: f(x) = sin(x), analytic derivative at x=0 is cos(0) = 1
f2 = lambda x: np.sin(x) 
numerical = numerical_deriv(f2, 0)
analytic = np.cos(0)
print(f"sin(x) at x = 0 | numerical: {numerical:.6f} | analytic: {analytic:.6f} | error: {abs(numerical - analytic):.2e}")

# test 3: f(x) = e^x, analytic derivative at x=1 is e^1 = 2.712...
f3 = lambda x: np.exp(x)
numerical = numerical_deriv(f3, 1)
analytic = np.exp(1)
print(f"e^x at x = 1 | numerical: {numerical:.6f} | analytic: {analytic:.6f} | error: {abs(numerical - analytic):.2e}")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
x = np.linspace(-3, 3, 500)

ax_left = axes[0]
y = x**3 - 2*x + 1
ax_left.plot(x, y, color='steelblue', linewidth=2)
ax_left.set_title("y = x^3 - 2x + 1")
ax_left.axhline(y=0, color='black', linewidth=0.7)
ax_left.axvline(x=0, color='black', linewidth=0.7)
ax_left.grid(True, alpha=0.3)

ax_right = axes[1]
f = lambda x: x**3 - 2*x + 1
fp = np.array([numerical_deriv(f, xi) for xi in x])
ax_right.plot(x, fp, color='darkred', linewidth=2)
ax_right.set_title('dy/dx')
ax_right.axhline(y=0, color='black', linewidth=0.7)
ax_right.axvline(x=0, color='black', linewidth=0.7)
ax_right.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
