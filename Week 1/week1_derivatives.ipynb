```python
import numpy as np


a = np.array([1, 2, 3, 4, 5])
b = np.zeros(5)
c = np.ones(5)
d = np.linspace(0, 10, 6)

print("From list:", a)
print("Zeroes:", b)
print("Ones:", c)
print("Linspace:", d)
```

```python
x = np.array([1, 2, 3, 4, 5])
print(x + 10)
print(x * 3)
print(x ** 2)
print(x + x)
```

```python
x = np.linspace(0, 2 * np.pi, 100)

print(np.sin(x[:5]))
print(np.exp(x[:5]))
print(np.sqrt(x[:5]))
print(np.log(x[1:6]))
```

```python
x = np.linspace(-5, 5, 500)

y = x**3 - 2*x + 1

print("x values (first 5):", x[:5])
print("y values (first 5):", x[:5])
print("shape of x:", x.shape)
print("shape of y:", y.shape)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
plt.plot(x, y, color='steelblue', linewidth=2)
plt.axhline(y=0, color='black', linewidth=0.8)
plt.axvline(x=0, color='black', linewidth=0.8)
plt.title("f(x) = x^3 - 2x + 1")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
